# Imports

```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD, Adam

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import time
```

# Carregando a base

```python
def load():
    data = pd.read_csv("diabetes.csv")
    X = data.iloc[:, :-1].values  
    z = data.iloc[:, 8].values
    lb = LabelEncoder()
    z[:] = lb.fit_transform(z[:])
    y = z.astype('int') 
    
    return X,y
```

```python
X.shape
```

```python
X, y = load()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
treino = []
acuracia = []
teste = []
```

# Shallow Model

```python
model = Sequential()
model.add(Dense(1, input_shape=(8,), activation='sigmoid'))
model.compile(Adam(lr=0.05), 'binary_crossentropy', metrics=['accuracy'])
```

```python
ti_treino = time.time()
model.fit(X_train, y_train, epochs=200, verbose=1)
tf_treino = time.time()

time_treino_shallow = tf_treino - ti_treino
treino.append(time_treino_shallow)
```

```python
ti_teste = time.time()
results=model.evaluate(X_test, y_test)
tf_teste = time.time()

time_teste_shallow = tf_teste - ti_teste

teste.append(time_teste_shallow)
acuracia.append(results[1])
```

```python
print('Train ACC: ', results[1])
print('Time Treino: ', time_treino_shallow)
print('Time Teste: ', time_teste_shallow)
```

# Deep Model

```python
deepModel = Sequential()
deepModel.add(Dense(4, input_shape=(8,), activation='tanh')) 
deepModel.add(Dense(2, input_shape=(16,), activation='tanh')) 
deepModel.add(Dense(1, input_shape=(8,), activation='sigmoid')) 
deepModel.compile(optimizer=Adam(lr=0.05), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
ti_treino = time.time()
deepModel.fit(X_train, y_train, epochs=200, verbose=1)
tf_treino = time.time()

time_treino_deep = tf_treino - ti_treino

treino.append(time_treino_deep)
```

```python
ti_teste = time.time()
deepResults = deepModel.evaluate(X_test, y_test)
tf_teste = time.time()

time_teste_deep = tf_teste - ti_teste

teste.append(time_teste_deep)
acuracia.append(deepResults[1])
```

```python
print('Train ACC: ', deepResults[1])
print('Time Treino: ', time_treino_deep)
print('Time Teste: ', time_teste_deep)
```

# Classificador QDA

```python
qda = QuadraticDiscriminantAnalysis(store_covariance=True)
tit_qda = time.time()
qda.fit(X, y)
tft_qda = time.time()
    
ti_qda = time.time()
qda_predicted = qda.predict(X_test)
tf_qda = time.time()
    
    #QDA 
print('----- Método QDA -----')
acc_qda = accuracy_score(y_test, qda_predicted)
acuracia.append(acc_qda)
print('Acurácia QDA :', acc_qda)
time_treino_qda = tft_qda - tit_qda
treino.append(time_treino_qda)
print('Time de Treino: ', time_treino_qda)
time_teste_qda = tf_qda - ti_qda
teste.append(time_teste_qda)
print('Time de Teste: ', time_teste_qda)
    
```

# Classificador MLP

```python
mlp = MLPClassifier(solver='adam', learning_rate='adaptive', hidden_layer_sizes = (100,), 
                        max_iter=9000, learning_rate_init=5e-04, tol=1e-4)

tit_mlp = time.time()
mlp.fit(X, y)
tft_mlp = time.time()
    
ti_mlp = time.time()
mlp_predicted = mlp.predict(X_test)
tf_mlp = time.time()
    
    
    #MLP
print('----- Método MLP -----')
acc_mlp = accuracy_score(y_test, mlp_predicted)
acuracia.append(acc_mlp)
print('Acurácia MLP: ', acc_mlp)
time_treino_mlp = tft_mlp - tit_mlp
treino.append(time_treino_mlp)
print('Time de Treino: ', time_treino_mlp)
time_teste_mlp = tf_mlp - ti_mlp
teste.append(time_teste_mlp)
print("Time de Teste: ", time_teste_mlp)
```

# Classificador KNN

```python
neigh = KNeighborsClassifier(n_neighbors=3)

tit_knn = time.time()
neigh.fit(X_train, y_train)
tft_knn = time.time()

ti_knn = time.time()
knn_predicted = neigh.predict(X_test)
tf_knn = time.time()

print('----- Método KNN -----')
acc_knn = accuracy_score(y_test, knn_predicted)
acuracia.append(acc_knn)
print('Acurácia MLP: ', acc_mlp)
time_treino_knn = tft_knn - tit_knn
treino.append(time_treino_knn)
print('Time de Treino: ', time_treino_knn)
time_teste_knn = tf_knn - ti_knn
teste.append(time_teste_knn)
print("Time de Teste: ", time_teste_knn)
```

# Tabela com os resultados

```python
columns = ['Classificador', 'Acurácia', 'Tempo de Treino', 'Tempo de Teste']
linha = ['Shallow Model', 'Deep Model', 'QDA', 'MLP', 'KNN']

df_linha = pd.DataFrame(data = linha, columns = ['Classificador'] )
df_acc = pd.DataFrame(data = acuracia, columns = ['Acurácia'])
df_treino = pd.DataFrame(data = treino, columns = ['Tempo de Treino'])
df_teste = pd.DataFrame(data = teste, columns = ['Tempo de Teste'])

pd.concat([df_linha, df_acc, df_treino, df_teste], axis = 1, join = 'inner')
```
